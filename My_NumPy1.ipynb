```python
# using linspace
# First argument is lower bound, second is upper bound and last is number of elemements which 
 # are linearly spaced....
my_series = linspace(0,100,24)

print(my_series)
```

```python

# In same block it will create the plot in the existing plot
my_x2 = array([2,3,4,5,6,7,8,9])
plot(my_x2)
my_y2 = array([3,4,5,3,2,2,4,6])
plot(my_y2)
```

```python
# If we want to plot the graph in seperate figures and inspite of writing the code in 
 # the same block
    
my_x2 = array([2,3,4,5,6,7,8,9])
my_y2 = array([3,4,5,3,2,2,4,6])

# This figure function will create the figure only and then on plotting it will
 # plot something....
figure()
plot(my_x2)


# Then calling the figure function will create seperate figure and will plot on it seperately
figure()
plot(my_y2)
    
```
